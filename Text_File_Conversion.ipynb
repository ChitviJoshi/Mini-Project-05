# Text File Conversion
####Compare the performance (time taken) of C, C++, Java, R, and Python programs for
####1. Convert 200MB, 400 MB, 600 MB, 800 MB, and 1000MB text files to upper case.

```python
import matplotlib.pyplot as plt

file_sizes = [200, 400, 600, 800, 1000]

c_times = [12, 20, 34, 45, 55]
cpp_times = [15, 25, 36, 50, 60]
java_times = [18, 30, 40, 55, 70]
r_times = [20, 35, 45, 60, 80]
python_times = [25, 40, 53, 75, 100]

plt.plot(file_sizes, c_times, marker='o', label='C', linewidth=2)
plt.plot(file_sizes, cpp_times, marker='s', label='C++', linewidth=2)
plt.plot(file_sizes, java_times, marker='^', label='Java', linewidth=2)
plt.plot(file_sizes, r_times, marker='d', label='R', linewidth=2)
plt.plot(file_sizes, python_times, marker='*', label='Python', linewidth=2)

# Chart styling
plt.title("Uppercase Conversion Performance Across Languages")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (seconds)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```
